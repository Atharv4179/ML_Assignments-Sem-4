```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv("C:/Users/chinc/Downloads/Admission_Predict_Ver1.1.csv")
```

```python
print(df.columns)
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['GRE Score'], y=df['CGPA'])
plt.title("Scatter Plot")
plt.xlabel("GRE Score")
plt.ylabel("CGPA")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=df['SOP'], y=df['CGPA'])
plt.title("Average CGPA for Different SOP Ratings")
plt.xlabel("SOP Rating")
plt.ylabel("Average CGPA")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['University Rating'], y=df['CGPA'])
plt.title("Box Plot: CGPA vs University Rating")
plt.xlabel("University Rating")
plt.ylabel("CGPA")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df.groupby('University Rating')['CGPA'].count().plot(kind='pie', autopct='%1.1f%%')
plt.title("Pie Chart: University Rating Distribution")
plt.ylabel("")  # Hide y-label
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(x=df['Serial No.'], y=df['CGPA'])
plt.title("Line Chart: CGPA Trend Over Serial Numbers")
plt.xlabel("Serial No.")
plt.ylabel("CGPA")
plt.show()
```
